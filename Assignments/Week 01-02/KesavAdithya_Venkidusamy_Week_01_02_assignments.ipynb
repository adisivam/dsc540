ACTIVITY 4: DESIGN YOUR OWN CSV PARSER <br>

A CSV file is something you will encounter a lot in your life as a data practitioner. A CSV is a comma-separated file where data from a tabular format is generally stored and separated using commas, although other characters can also be used. <br>

In this activity, we will be tasked with building our own CSV reader and parser. Although it is a big task if we try to cover all use cases and edge cases, along with escape characters and all, for the sake of this small activity, we will keep our requirements small. We will assume that there is no escape character, meaning that if you use a comma at any place in your row, it means you are starting a new column. We will also assume that the only function we are interested in is to be able to read a CSV file line by line where each read will generate a new dict with the column names as keys and row names as values. <br>

Here is an example:

```python
#Importing lib required for this exercise
from itertools import zip_longest
import re
```

```python
#Function to contruct the dictionary
def dict_constructor_fn(header, line):
    ''' dictionary constructor function to take in header and data elements and construct the dictionary and return it'''
    zip_out = zip_longest(header, line, fillvalue=None)       
    dct = {d1[0] : d1[1] for d1 in zip_out}
    return dct
```

```python
#Code snippet to read csv and create output csv parser
output = []
dict_csv = {}

with open("sales_record.csv",'r') as f:
    content = f.readlines()
f.close()

h = 1
for i in content:
    if h == 1:
        header = (re.sub(r'[\n]', r'', i)).split(',')
        h = 2
    else:
        line = (re.sub(r'[\n]', r'', i)).split(',')
        dict_csv = dict_constructor_fn(header,line)
        output.append(dict_csv)
```
