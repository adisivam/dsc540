```python
# 1. Load the necessary libraries.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 2. Read in the Boston housing dataset (given as a .csv file) from the local directory.

housing_df = pd.read_csv("Boston_housing.csv")
```

```python
# 4. Create a smaller DataFrame with columns that do not include CHAS, NOX, B, and LSTAT.

cols_list = ['CRIM', 'ZN', 'INDUS', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'PRICE']

housing_subset_df = housing_df[cols_list]

housing_subset_df.head()
```

```python
# 5. Check the last seven records of the new DataFrame you just created.

housing_subset_df.tail(7)
```

```python
# 8. Create a scatter plot of crime rate versus price.

housing_df.plot.scatter('CRIM', 'PRICE', edgecolor='b')
plt.grid('True')
plt.title("Crime Rate vs. House Prices plot", fontsize=12)
plt.xlabel("Crime Rate", fontsize=12)
plt.ylabel("House Prices", fontsize=12)
plt.show()
```

```python
# 9. Plot using log10(crime) versus price.

x = np.log10(housing_df.CRIM)
y = housing_df['PRICE']

plt.scatter(x, y)
plt.grid('True')
plt.title("log10 of Crime Rate vs. House Prices plot", fontsize=12)
plt.xlabel("log10 of Crime Rate", fontsize=12)
plt.ylabel("House Prices", fontsize=12)
plt.show()
```

```python
# 10. Calculate some useful statistics, such as mean rooms per dwelling, median age, 
#     mean distances to five Boston employment centers, and the percentage of houses with a low price (< $20,000)

print("Mean Rooms per dwellings is: {:.2f}".format(housing_subset_df.RM.mean()))
print("Median age is: {:.2f}".format(housing_subset_df.AGE.median()))
print("Mean Distance to employment centers is: {:.2f}".format(housing_subset_df.DIS.mean()))

fraction_houses = ( housing_subset_df[housing_subset_df.PRICE < 20].PRICE.count() / housing_subset_df.PRICE.count() ) * 100
print("Percentage of houses with a low price below $20,000 is: {:.2f}".format(fraction_houses))
```

CSV file doesn't have header row. So, header row needs to be derived from adult income names text file

```python
# 3. Create a script that will read a text file line by line and create header row for the dataframe
# 4. Add a name of Income for the response variable to the dataset.

header = []
with open('adult_income_names.txt', 'r') as fd:
    for line in fd:
        if line != '\n':
            x = line.split(':')[0]
            header.append(x)

header.append("income")
print("Header row derived from adult income names files")
print(header)
        
```

```python
# 4. Read the csv file along with header

adult_income_df = pd.read_csv("adult_income_data.csv", names=header)
adult_income_df.head(3)
```

```python
# 5. Find the missing values.

adult_income_df.isnull().sum()
```

```python
# 6. Create a DataFrame with only age, education, and occupation by using subsetting.

adult_income_subset_df = adult_income_df[['age','education','occupation']]
adult_income_subset_df.head(5)
```

```python
# 7. Plot a histogram of age with a bin size of 20.

adult_income_subset_df.age.hist(bins=20)
plt.title(f'Histogram for Age')
```

```python
# 8. Create a function to strip the whitespace characters.

def strip_whitespace(in_value):
    return in_value.strip()
```

```python
# 9. Use the apply method to apply this function to all the columns with string values, create a new column, 
#    copy the values from this new column to the old column, and drop the new column.

print("The data type of each column present in the dataframe")
print(adult_income_df.dtypes)
adult_income_df.head(3)
```

```python
# Looping through all the string columns and apply strip whitespace function

str_cols = ['workclass','education','marital-status','occupation','relationship','sex','native-country']

for col in str_cols:
    col_temp = col + '_temp'
    adult_income_df[col_temp] = adult_income_df[col].apply(strip_whitespace)
    adult_income_df[col] = adult_income_df[col_temp]
    adult_income_df.drop(col_temp,axis=1,inplace=True)
    
```

```python
# Printing the dataframe with all columns after applying strip whitespaces

adult_income_df.head(5)
```

```python
# 10. Find the number of people who are aged between 30 and 50.

cnt = adult_income_df[(adult_income_df.age >= 30) & (adult_income_df.age <= 50)].age.count()
print('Total number of people aged between 30 and 50 is: {}'.format(cnt))
```

```python
# 11. Group the records based on age and education to find how the mean age is distributed.

adult_income_subset_df.groupby('education').describe()
```

```python
adult_income_subset_df.boxplot(column='age', by='education', figsize=(35, 15))
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('education', fontsize=20)
plt.show()
```

```python
# 12. Group by occupation and show the summary statistics of age. Find which profession has the oldest workers on average 
#     and which profession has its largest share of the workforce above the 75th percentile.

byOccupation = adult_income_subset_df.groupby('occupation').describe()['age']
byOccupation
```

```python
# Profession has the oldest workers on average

byOccupation[byOccupation['mean'] == max(byOccupation['mean'])]
```

```python
# which profession has its largest share of the workforce above the 75th percentile.

occupation_75_df = adult_income_subset_df[adult_income_subset_df['age'] > np.percentile(adult_income_subset_df['age'], 75)].groupby('occupation').describe()['age']
occupation_75_df
```

```python
print('Profession has its largest share of the workforce above the 75th percentile: \n')
occupation_75_df[occupation_75_df['count'] == max(occupation_75_df['count'])]
```

```python
# 13. Use subset and groupby to find outliers.

#occupation_stats= df_subset.groupby('occupation').describe()['age']
byOccupation = adult_income_df[['occupation','age']].groupby('occupation').describe()['age']
byOccupation
```

##### By looking at the descriptive stats, we could see the count for "Armed-Force" occupation is only 9 which could be outlier in this case. This can be further confirmed by plotting a bar chart for occupation. 

```python
# 14. Plot the values on a bar chart.

plt.figure(figsize=(15,8))
plt.barh(y=byOccupation.index,width=byOccupation['count'])
plt.yticks(fontsize=13)
plt.show()
```

```python
# 15. Merge the data using common keys.

income_subset_df1 = adult_income_df[['age','workclass','occupation']].sample(5,random_state=101)
income_subset_df1.head()
```

```python
income_subset_df2 = adult_income_df[['education','income','occupation']].sample(5,random_state=101)
income_subset_df2.head()
```

```python
df_merged = pd.merge(income_subset_df1,income_subset_df2,on='occupation',how='inner').drop_duplicates()
df_merged
```

#### 3. Create a series and practice basic arithmetic steps

a. Series 1 = 7.3, -2.5, 3.4, 1.5 <br>
    i. Index = a, c, d, e <br>
b. Series 2 = -2.1, 3.6, -1.5, 4, 3.1 <br>
    i. Index = a, c, e, f, g <br>
c. Add Series 1 and Series 2 together and print the results <br>
d. Subtract Series 1 from Series 2 and print the results

```python
# Declaring the elements
idx_1  = ['a', 'c', 'd', 'e']
list_1 = [7.3, -2.5, 3.4, 1.5]

idx_2  = ['a', 'c', 'e', 'f', 'g']
list_2 = [-2.1, 3.6, -1.5, 4.0, 3.1]
```

```python
# Declaring the series
series_1 = pd.Series(data=list_1, index=idx_1)
series_2 = pd.Series(data=list_2, index=idx_2)
```

```python
# Addition and Subtraction of series
series_add   = series_1 + series_2
series_sub = series_2 - series_1
```

```python
print("Addition of series: series_1 + series_2")
print(series_add)
print('\n')
print('Subtraction of series: series_2 - series_1')
print(series_sub)
```
