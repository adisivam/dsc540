#### Transformation 2: Generate date range

Since the data available for 7 days (2016-10-24 thru 2016-10-30) in the dataset, we will generate Hour range and minute range index. Starting point would be the Timestamp from the first and last occurrences from 'Timestamp' column - the data is already sorted on the Timestamp column.

```python
## Create Date range on Hour basis using the Timestamps from '0' occurrence and '-1' (last) occurrence from DataFrame

hour_range = pd.date_range(candy_hier_df['Timestamp'].iloc[0], candy_hier_df['Timestamp'].iloc[-1], freq='H')
```

```python
## Display the hour_range index generated

hour_range
```

```python
## Create Date range on minutes basis using the Timestamps from '0' occurrence and '-1' (last) occurrence from DataFrame

minute_range = pd.date_range(candy_hier_df['Timestamp'].iloc[0], candy_hier_df['Timestamp'].iloc[-1], freq='T')
```

```python
minute_range
```
